```python
%config Completer.use_jedi = False

import configparser
config = configparser.ConfigParser()
config.read('har-config.txt')

source_folder = config.get('Configuration', 'source_folder')
raw_folder = config.get('Configuration', 'raw_folder')

time_step = int(config.get('Configuration', 'time_step'))
time_step2 = int(config.get('Configuration', 'time_step2'))
window_size = int(config.get('Configuration', 'window_size'))
data_folder = config.get('Configuration', 'data_folder')
ini_date = config.get('Configuration', 'ini_date')
end_date = config.get('Configuration', 'end_date')


w = int(config.get('Configuration', 'w'))
h = int(config.get('Configuration', 'h'))


rooms=config.get('Configuration', 'rooms')
rooms = [(item.strip()) for item in rooms.split(',')]

```

```python
import numpy as np
from pandas import read_csv    
import pandas as pd
import time

from datetime import datetime

off_zone=60*60*2

def day_time(ti):
    return int((int)((ti+off_zone)/(60*60*24)))


def time2str(tt):
    ttt=int(tt)
    return datetime.fromtimestamp(ttt).strftime("%Y-%m-%d %H:%M:%S")

tN = (int)(datetime.strptime(end_date, '%Y-%m-%d %H:%M:%S').timestamp())
print (end_date,"End date is", tN, "day:",day_time(tN), time2str(tN))


print(ini_date)
t0 = (int)(datetime.strptime(ini_date, '%Y-%m-%d %H:%M:%S').timestamp())
print (ini_date,"Init date is", t0, "day:",day_time(t0), time2str(t0))


def absT(tt):
    return (int)(tt*time_step+t0)

def day_time0(ti):
    return day_time(ti)-day_time(t0)+1

def relT(ti):
    return (int)((ti-t0)/time_step)



def sizeT():
    return relT(tN) 

day=list(range(day_time(t0),day_time(tN)+1))[0]
print(day_time(t0))

print(t0,time2str((t0)))


from datetime import datetime, timedelta


def getStrDatefrom(day_number):
    # Reference date (1970-01-01)
    reference_date = datetime(1970, 1, 1)
    
    # Compute the date corresponding to the day number
    resulting_date = reference_date + timedelta(days=day_number)
    
    return resulting_date.strftime("%Y-%m-%d")




day_str=getStrDatefrom(day)
print(day_str)
users=config.get('Configuration', day_str)
users = [(item.strip()) for item in users.split(',')]


print(users)
```

```python
def getLoc(room,s,off):
    print("S:",s)
    xys=config.get(room, s)
    xys = [(item.strip()) for item in xys.split(',')]
    xys = [list(map(int, item.split('x'))) for item in xys]
    for xy in xys:
        print("\t",xy)
        xy[0]=xy[0]+off[0]
        xy[1]=xy[1]+off[1]
        print("\t+",xy)
    return xys

loc_room_sensor={}
obj_sensors={}
continous_sensors={}
for room in rooms:
    print(room)
    obj_sensors[room]={}
    offW = int(config.get(room, 'offW'))
    offH = int(config.get(room, 'offH'))
    try:
        obj_sensors[room] = config.get(room, 'object_sensors')
        obj_sensors[room]  = [(item.strip()) for item in obj_sensors[room].split(',')]


        for obj in obj_sensors[room]:
            print("\t obj",obj)
            loc_room_sensor[room][obj]=getLoc(room,obj,(offH,offW))
            print(obj_sensors[room][obj])
    except Exception as e:
        print("not object_sensors in ",room,e)
        
    continous_sensors[room]={}
    try:
        continous_sensors[room] = config.get(room, 'continous_sensors')
        continous_sensors[room]  = [(item.strip()) for item in continous_sensors[room].split(',')]

        for obj in continous_sensors[room]:
            print("\t cont obj",obj)
            loc_room_sensor[room][obj]=getLoc(room,obj,(offH,offW))
            print(obj_sensors[room][obj])
    except Exception as e:
        print("not object_sensors in ",room,e)
print(loc_room_sensor)
print(obj_sensors)
```

```python
import cv2
import os


os.makedirs(data_folder+"DAY_"+str(day)+"/object_sensor/", exist_ok=True)


def valContinuous(sensor):
    ty=config.get(sensor, 'type')
    
    def relative_change(stream,ti,tlong,tshort):
        tlong2=max(ti - tlong, 0)
        tshort2=max(ti - tshort, 0)

        print("\t",ti,time2str(absT(ti)),tlong2,tshort2,np.mean(stream[tlong2:ti]),np.mean(stream[tshort2:ti]),stream[tlong2:ti])
        return np.mean(stream[tlong2:ti])-np.mean(stream[tshort2:ti])

    def absolute_value(stream,ti,time_step2):
        tshort2=max(ti - time_step2, 0)
        return np.mean(stream[tshort2:ti])
    
    def max_timestep(stream,ti,time_step2):
        return np.max(stream[ti:ti+time_step2])
    
    if ty == "relative_change":
        parametros_por_defecto = {
        'tlong': int(config.get(sensor, 'avg_window')),
        'tshort': int(config.get(sensor, 'current_window'))
        }
    
        return relative_change, parametros_por_defecto
    if ty == "absolute_value":
        parametros_por_defecto = {
            'time_step2':int(time_step2)
        }    
        return absolute_value, parametros_por_defecto
    if ty == "max_timestep":
        parametros_por_defecto = {
            'time_step2':int(time_step2)
        }    
        return absolute_value, parametros_por_defecto
    return None,None
        
for room in rooms:
    for obj in list(obj_sensors[room]) + list(continous_sensors[room]):
        with open(data_folder+"/DAY_"+str(day)+"/object_sensor/"+obj+".tsv", "w") as file: 
            
            fun, param=valContinuous(obj)
            try:
                condition=config.get(obj, 'condition')
            except Exception as e:
                condition=obj+">0"
            print(obj,fun,param, condition)
            
            for tt in range(t0,tN,time_step2):
                    #print(tt,day_time(tt),time2str(tt))
                    ti=relT(tt)
                    #print(tt,ti)
                    v=fun(df_total_map2[room][obj],ti+time_step2,*param.values())
                    expression = condition.replace(obj, str(v))
                    ok=1 if eval(expression) else 0

                    #print(tt,obj,v,ok)
                    file.write(time2str(tt)+"\t"+str(ok)+"\n")
```
